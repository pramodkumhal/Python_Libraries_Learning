#Series

> It is defined as a one-dimensional array taht is capable of storing various data types


````python
#Creating Series with Custom index and datatype
import pandas as pd
x=[3,4,5,6]
var=pd.Series(x,index=['a','v','f','h'],dtype='float',name='python')
print(var)
print(type(var))
print(var[3])
````

````python
#Using Series from dictonary
dic={'name':['python','c','c++','java'],'por':[12,13,14],'rank':[1,4,3,2]}
var1=pd.Series(dic)
print(var1)
````

````python
s=pd.Series(12,index=[1,2,3,4,5,6])
print(s)
print(type(s))
````

````python
#There is no broadcast error
s1=pd.Series(12,index=[1,2,3,4,5,6,7])
s2=pd.Series(12,index=[1,2,3,4])
print(s1 + s2)
````

#Arthmetic Operation

> Addition, Subtraction, Multiply and Divide



````python
import pandas as pd
var = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
print(var)
````

````python
#Addtion
var['C']=var['A']+var['B']
print(var)
#Subtraction
var['C']=var['A']-var['B']
print(var)
#Multiplication
var['C']=var['A']*var['B']
print(var)
#Divide
var['C']=var['A']/var['B']
print(var)
````

````python
#Printing only specific data
var1 = pd.DataFrame({'A':[10,20,30,40],'B':[15,16,17,18]})
print(var1)
print()
var1['Python']=var1['A'] <= 20
var1['Python_1']=var1['B'] >= 16
print(var1)
````

#Insert and Delete

````python
import pandas as pd
var = pd.DataFrame({'A':[10,20,30,40],'B':[15,16,17,18]})
print(var)
print()
var.insert(1,'Python',var['A'])
print(var)
print()
var['python_1'] = var['A'][:3]
print(var)
print()
#Delete
del var['A']
print(var)
````

#**Merging and Concat**

````python
#Merge
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'A':[1,2,3,5],'C':[21,22,23,24]})
#pd.merge(var1,var2,how='outer', indicator=True)
````

````python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]})
pd.merge(var1,var2,left_index=True, right_index=True,suffixes=('name','python'))
````

````python
 #concat
 s1=pd.Series([1,2,3,4])
 s2=pd.Series([11,21,31,41])
 pd.concat([s1,s2])
````

````python
d1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
d2=pd.DataFrame({'A':[1,2,3,4],'C':[21,22,23,24]})
pd.concat([d1,d2],axis=1,keys=['d1','d2'])
````

#Groupby

````python
import pandas as pd
var=pd.DataFrame({'name':['a','b','c','d','a','b','a','b','a','c','c','d'],
                  'S_1':[12,13,14,12,13,14,15,23,25,16,10,34],
                  'S_2':[23,24,25,26,27,28,29,30,25,34,35,56]})
var
````

````python
var_new=var.groupby('name')
var_new

for x,y in var_new:
  print(x)
  print(y)
  print()
````

````python
#var_new.get_group('a')
print(var_new.max())
print()
print(var_new.mean())
print()
#print(list(var_new))
````
