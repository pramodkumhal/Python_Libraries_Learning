# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]

x_1 = pd.DataFrame({'var': var, 'var_1': var_1})

sns.lineplot(x='var', y='var_1', data=x_1)

plt.show()
```

# HeatMap Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var
```

```python
#Simple Heatmap
sns.heatmap(var)
plt.show()
```

```python
y={'fontsize':15,'color':'r'}
v=sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=4,
            linecolor='y',cbar=False)
v.set(xlabel='X-axis',ylabel='X-axis')  #function for label
sns.set(font_scale=0.5)   #function to control label fontsize
plt.show()
```

# Box Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
var=sns.load_dataset("tips")
var
```

```python
sns.set(style='whitegrid')
sns.boxplot(x='day',y='total_bill',data=var,order=['Fri','Sun','Thur','Sat'],
            showmeans=True,meanprops={'marker':'+','markeredgecolor':'black'},
            linewidth=3,palette='plasma')
plt.show()
```

# Styling Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
var=sns.load_dataset('tips')
var
```

```python
sns.set_style('darkgrid') #For styling Grid
sns.barplot(x='day',y='total_bill',data=var)
plt.show()
```

```python
sns.set_style('white')
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()   # Remove top and right spines
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(12,5))   #Changing size of figure
sns.barplot(x='day',y='total_bill',data=var)
sns.despine()
plt.show()
```

```python
sns.set_style('white')

sns.set_context('paper',font_scale=1)  #Set context for presentations

sns.barplot(x='day',y='total_bill',data=var)
plt.show()
```

# Multiple Plot (Facet-Grid)

```python
import seaborn as sns
import matplotlib.pyplot as plt
var=sns.load_dataset('tips')
var
```

```python
fg=sns.FacetGrid(var,col='day',hue='sex')
fg.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```
