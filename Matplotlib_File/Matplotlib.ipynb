# **Matplotlib**

```python
import matplotlib.pyplot as plt
```

# Bar Plot

```python

x=['Python','C','C++','Java']
y=[85,70,60,82]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("Bar Plot",fontsize=25)
c=["y",'b','m','g']

plt.bar(x,y,width=0.4,color='r',label='Popularity')

plt.legend()
plt.show()
```

```python
#Stacked Bar Plot
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("Bar Plot",fontsize=25)

plt.bar(x,y,width=0.4,color='r',label='Popularity')
plt.bar(x,z,width=0.4,color='y',label='Popularity1')

plt.legend()
plt.show()
```

```python
#Grouped Bar Plot
import matplotlib.pyplot as plt
import numpy as np

x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Language",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.title("Bar Plot",fontsize=25)

plt.bar(p,y,width,color='r',label='Popularity')
plt.bar(p1,z,width,color='y',label='Popularity1')

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
#Horizontal Bar
x=['Python','C','C++','Java']
y=[85,70,60,82]
z=[20,30,40,50]


plt.ylabel("Language",fontsize=15)
plt.xlabel("Number",fontsize=15)
plt.title("Bar Plot",fontsize=25)

plt.barh(x,y,color='r',label='Popularity') #barh is used for horizontal
plt.barh(x,z,color='y',label='Popularity1')

plt.legend()
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,4,5,3,6]

plt.scatter(day,number,color='r')

plt.title("Scatter Plot",fontsize=20)
plt.xlabel("day",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,4,5,3,6]
color=['r','g','y','b','r','g','r']
size=[100,150,180,140,130,90,70]
plt.scatter(day,number,c=color,s=size,)

plt.title("Scatter Plot",fontsize=20)
plt.xlabel("day",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,4,5,3,6]
number2=[3,2,1,4,5,6,2]
color=[10,49,30,29,56,20,30] #color['r','y','g','b','r','g','y'] we can also use this for multiple color
size=[100,150,180,140,130,90,70] #for colormap i.e cmap()

plt.scatter(day,number,c=color,s=size,cmap='viridis')
plt.scatter(day,number2,color='r',s=size)

t=plt.colorbar()  #showing color bar
t.set_label("ColorBar",fontsize=15)

plt.title("Scatter Plot",fontsize=20)
plt.xlabel("day",fontsize=15)
plt.ylabel("Number",fontsize=15)

plt.show()
```

# Pie Plot / Chart

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=['C','C++','Java','Python']

plt.pie(x,labels=y)
plt.title("Basic Pie Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=['C','C++','Java','Python']
ex=[0.3,0.0,0.0,0.0]
c=['red','blue','yellow','orange']

plt.pie(x,labels=y,explode=ex,colors=c,autopct='%0.1f%%',shadow=True,radius=0.8,
        labeldistance=1.2,startangle=0,textprops={'fontsize':12},counterclock=True,
        wedgeprops={'linewidth':5},rotatelabels=False,)
plt.title("Pie Chart / Plot")
plt.legend(loc=1)
plt.show()
```

```python
plt.pie([1])    #this 1 means radius of pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[40,30,20,10]
y=['C','C++','Java','Python']
c=['r','b','g','y']

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=0.5,colors=c)

plt.show()
```

# Steam Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt="g--",markerfmt='ro',basefmt='g',label='Python',use_line_collection=False)
plt.legend()
plt.show()
```

# Box Plot or Whisker Plot

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70]

plt.boxplot(x)
plt.title("Basic Box Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,70,120]
x1=[10,20,40,50,30,60,90,150]
y=[x,x1]
plt.boxplot(y,labels=['python','c++'],showmeans=True,sym='g',boxprops=dict(color='r'),capprops=dict(color='r'),
            whiskerprops=dict(color='r'),flierprops=dict(marker='+',markeredgecolor='r'))
plt.show()
```

# Stack Plot or Area Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[1,4,1,3,2]

plt.stackplot(x,area1,area2)
plt.title("Basic Stack Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area1=[2,3,2,5,4]
area2=[2,3,4,5,6]
area3=[1,3,2,4,2]

c=['r','g','m']
l=['area1','area2','area3']

plt.stackplot(x,area1,area2,area3,colors=c,labels=l,baseline='zero',edgecolor='black')
plt.title("Stack Plot",fontsize=20)
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-label",fontsize=15)
plt.legend(loc=2)
plt.show()
```

# Step Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,2,3,5,4]
plt.step(x,y,color='r',marker='o',ms=10,mfc='g',label='python',linestyle='--')
plt.title("Step Plot",fontsize=20)
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-label",fontsize=15)
plt.grid()
plt.legend(loc=2)
plt.show()
```

# Fill_Between Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')

plt.fill_between(x,y)

plt.title("Basic fill_bracket Plot")
plt.xlabel('X-axis')
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='g',where=((x>=2) & (x<=4)))

plt.title("fill_bracket Plot")
plt.xlabel('X-axis')
plt.ylabel("Y-axis")
plt.show()
```
